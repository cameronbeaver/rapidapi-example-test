```python
import mechanicalsoup
#import pandas as pd
import sqlite3

browser = mechanicalsoup.StatefulBrowser()

url = "https://www.macrotrends.net/1476/copper-prices-historical-chart-data"

browser.open(url)
print(browser.get_url())
```

```python
td = browser.page.find_all("td")
td

distribution = [value.text.replace("\n","") for value in td]
print(distribution.index("Copper Prices - 45 Year Historical Chart"))
distribution[:441]
```

```python
with open("copper.txt","w") as f:
    f.writelines(distribution)
```
